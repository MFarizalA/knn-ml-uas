```python
import operator
from operator import itemgetter
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
import os, sys
import cv2
```

```python
X = []
y = []
count = 0
dir="archive/105_classes_pins_dataset/"
for i in os.listdir(dir):
    count+=len(os.listdir(dir+"/"+i))
    for j in os.listdir(dir+"/"+i):
        img = cv2.imread(dir+"/"+i+"/"+j)
        img = cv2.resize(img, (128, 128))
        X.append(np.asarray(img))
        y.append(i)
X = np.asarray(X)
y = np.asarray(y)
print(X.shape, y.shape)
```

```python
X = X / 255
```

```python
X = X.reshape(17534, 16384).astype('float32')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
def euclidean_dist(x1, x2):
        return np.sqrt(np.sum((x1-x2)**2))
```

```python
class KNN:
    def __init__(self, K):
        self.K = K
    def fit(self, x_train, y_train):
        self.X_train = x_train
        self.Y_train = y_train
    def predict(self, X_test):
        predictions = []
        count = 0
        for i in range(len(X_test)):   
            count += 1         
            distance = np.array([euclidean_dist(X_test[i], x_train) for x_train in self.X_train])
            distance_sorted = distance.argsort()[:self.K]
            neighbor_count = {}
            for index in distance_sorted:
                if self.Y_train[index] in neighbor_count:
                    neighbor_count[self.Y_train[index]] += 1
                else:
                    neighbor_count[self.Y_train[index]] = 1       
            sorted_neigh_count = sorted(neighbor_count.items(), key=operator.itemgetter(1), reverse=True)
            print(str(count)+' '+str(sorted_neigh_count[0][0]))
            predictions.append(sorted_neigh_count[0][0])
        return predictions
```

```python
accs = []
k = 3
model = KNN(K = k)
model.fit(X_train, y_train)
pred = model.predict(X_test)
acc = accuracy_score(y_test, pred)
accs.append(acc)
```
